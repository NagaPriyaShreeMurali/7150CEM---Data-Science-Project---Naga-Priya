### Extracting respondents with age from 20-60 years old

```python
df_er = df_pd[(df_pd['AgeInYrs'] >= 20) & (df_pd['AgeInYrs'] <= 60)]
df_er
```

```python
#Creating a new column to define obesity level for each respondent data
# 1 - Under Weight
# 2 - Healthy
# 3 - Overweight
# 4 - Obese
```

```python
obese_lvl = [(df_er['BMI'] < 18.5),
                   (df_er['BMI'] >= 18.5) & (df_er['BMI'] < 25.0),
                   (df_er['BMI'] >= 25.0) & (df_er['BMI'] < 30.0),
                   (df_er['BMI'] > 30.0)]
val = [1, 2, 3, 4]

df_er['obesityLevel'] = np.select(obese_lvl, val, default = 1)
df_er
```

### Checking if there are missing values

```python
df_er.isnull().sum()
```

### Fill in NaN with mean,mode values

```python
df_er['IncomeRatio'].fillna(df_er['IncomeRatio'].mean(), inplace=True)
df_er['SleepHours-Workdays'].fillna(df_er['SleepHours-Workdays'].mean(), inplace=True)
df_er['SleepHours-Weekend'].fillna(df_er['SleepHours-Weekend'].mean(),inplace=True)
```

```python
mode_PoorAppetiteOvereating = df_er['PoorAppetiteOvereating'].mode()[0]
df_er['PoorAppetiteOvereating'].fillna(mode_PoorAppetiteOvereating,inplace=True)

mode_Depressed = df_er['Depressed'].mode()[0]
df_er['Depressed'].fillna(mode_Depressed,inplace=True)

mode_Smoking = df_er['Smoking'].mode()[0]
df_er['Smoking'].fillna(mode_Smoking,inplace=True)
```

### Normalizing Data

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0,1))
normalized_data = scaler.fit_transform(df_er)
columns = ['Gender','AgeInYrs','Origin','BirthCountry','EducationLevel','IncomeRatio','HeightInCm','Diabetes','ModerateWorkActivity',
           'ModerateRecreationalActivities','Depressed','PoorAppetiteOvereating','Smoking','SleepHours-Workdays',
           'SleepHours-Weekend', 'obesity2']
final_df = pd.DataFrame(normalized_data, columns=columns)
final_df['obesity2'] = final_df['obesity2'].astype(int)
final_df.head()
```

# Test and train data

```python
from sklearn.model_selection import train_test_split
X = final_df.iloc[:, :-1]
Y = final_df.iloc[:, -1]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

# Baseline model
A DummyClassifier is a classifier that generates predictions based on basic rules. It serves as a baseline model for comparison with other,more complex models.

```python
from sklearn.dummy import DummyClassifier
baseline_classifier = DummyClassifier(strategy = "most_frequent")
baseline_classifier.fit(X_train,Y_train)
# predicting
Y_pred_base = baseline_classifier.predict(X_test)
# accuracy calculation
from sklearn import metrics
print("Accuracy:", metrics.accuracy_score(Y_test, Y_pred_base))
```

+ The initial accuracy is at 57.64%, showing the lowest level of prediction we can achieve. We're hoping that the models we've chosen will deliver much better accuracy.

# Random Forest 

```python
# Random Forest Model
rand_forest = RandomForestClassifier(random_state=42)
rand_forest.fit(X_train,Y_train)

Y_pred_rf = rand_forest.predict(X_test)
print("Accuracy:", metrics.accuracy_score(Y_test, Y_pred_rf))
```

```python
# cross validation
from sklearn.model_selection import cross_val_score
cv_rf = cross_val_score(rand_forest, X, Y, scoring='accuracy', cv=10)
cv_rf = pd.Series(cv_rf)
cv_rf.mean()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

# Define a smaller parameter grid to start with
param_grid = {
    'n_estimators': [100, 200],  # Fewer options for the number of trees
    'max_depth': [10, 20]  # Fewer options for maximum depth
}

# Create a base model
rf = RandomForestClassifier(random_state=42)

# Instantiate the grid search model
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, 
                           cv=5, n_jobs=-1, verbose=2, scoring='accuracy')

# Fit the grid search to the data
grid_search.fit(X_train, Y_train)
# Print the best parameters
#print("Best Parameters Found: ", grid_search.best_params_)

# Use the best parameters to fit the model
best_grid = grid_search.best_estimator_

# Make predictions using the model fitted with the best parameters
Y_pred_best = best_grid.predict(X_test)

# Calculate accuracy
print("Accuracy: ", metrics.accuracy_score(Y_test, Y_pred_best))
```

### Precision-Recall Curves

```python
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, auc

# Rerun Models and Get the Predict Probabilities
rf_alg = rand_forest.fit(X_train, Y_train)
probs_rf = rf_alg.predict_proba(X_test)[:, 1]  # Probabilities for Random Forest

nb_alg = nb_clf.fit(X_train, Y_train)
probs_nb = nb_alg.predict_proba(X_test)[:, 1]  # Probabilities for Naive Bayes

xgb_alg = XGB_Model.fit(X_train, Y_train)
probs_xgb = xgb_alg.predict_proba(X_test)[:, 1]  # Probabilities for XGBoost

lg_alg = log_reg.fit(X_train,Y_train)
probs_lr = lg_alg.predict_proba(X_test)[:, 1]  # Probabilities for Logistic Regression

# Calculate precision and recall for each model
precision_rf, recall_rf, _ = precision_recall_curve(Y_test, probs_rf)
rf_prs = auc(recall_rf, precision_rf)

precision_nb, recall_nb, _ = precision_recall_curve(Y_test, probs_nb)
nb_prs = auc(recall_nb, precision_nb)

precision_xgb, recall_xgb, _ = precision_recall_curve(Y_test, probs_xgb)
xgb_prs = auc(recall_xgb, precision_xgb)

precision_lr, recall_lr, _ = precision_recall_curve(Y_test, probs_lr)
lr_prs = auc(recall_lr, precision_lr)

# Plot the Precision-Recall Curves for each model
plt.figure(figsize=(12, 8))
plt.plot(recall_rf, precision_rf, label=f'Random Forest (AUC = {rf_prs:.2f})')
plt.plot(recall_nb, precision_nb, label=f'Naive Bayes (AUC = {nb_prs:.2f})')
plt.plot(recall_xgb, precision_xgb, label=f'XGBoost (AUC = {xgb_prs:.2f})')
plt.plot(recall_lr, precision_lr, label=f'Logistic Regression (AUC = {lr_prs:.2f})')

plt.title('Precision-Recall Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend()
plt.show()
```

```python
# Print AUC for each model
print(f'AUC of Random Forest: {rf_prs:.2f}')
print(f'AUC of Naive Bayes: {nb_prs:.2f}')
print(f'AUC of XGBoost: {xgb_prs:.2f}')
print(f'AUC of Logistic Regression: {lr_prs:.2f}')
```

+ The Random Forest model has the highest AUC of 0.59, which suggests it has the best trade-off between Precision and Recall across different thresholds among the models shown.
